Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States
Assualt- Assualt rate in different places of United States
UrbanPop - urban population in different places of United States
Rape - Rape rate in different places of United States


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
```

```python
crime = pd.read_csv("/content/crime_data.csv")
crime
```

```python
crime.info()
```

```python
# Normalize heterogenous numerical data using z-score (x-mean/std) or custom defined function 
# Normalization function - here custom defined
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(crime.iloc[:,1:])
crime_norm
```

# **Hierarchial Clustering**

```python
# Create Dendrograms 
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(crime_norm,'average'))
```

```python
# Create Clusters (y)
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
y_hc=hc.fit_predict(crime_norm)
y_hc
```

```python
crime['h_clusterid']=hc.labels_
crime
```

```python
crime.groupby('h_clusterid').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 14))  
plt.scatter(crime['h_clusterid'],crime['City'], c=hc.labels_)
```

# **K Means Clustering**

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(crime_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(crime_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime['Clust']=x
crime
```

```python
crime.iloc[:,1:5].groupby(crime.Clust).mean()
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=crime.iloc[:,1:5]
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
c1=pd.DataFrame(dbscan.labels_,columns=['cluster'])
c1
```

```python
pd.concat([crime,c1],axis=1)
```
